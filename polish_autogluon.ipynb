```python
import pandas as pd
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
df = pd.read_csv('polish_train_split.csv')

target = 'company_bankrupt'
```

```python
predictor = TabularPredictor(label=target).fit(
    df,
    presets='high_quality',
    time_limit=3600
)
```

```python
test_df = pd.read_csv("polish_test_split.csv")

test_features = test_df.drop(columns=[target])
predictions = predictor.predict(test_features)
```

```python
predictor.predict_proba(test_features)
```

```python
predictor.evaluate(test_df)
```

```python
lb = predictor.leaderboard(silent=True)
print(lb)
```
